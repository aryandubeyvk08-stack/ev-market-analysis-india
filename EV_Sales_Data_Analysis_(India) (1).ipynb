```python
import pandas as pd

df = pd.read_csv("final_dataset.csv")

print(df.head())
print(df.columns)
print(df.info())
```

```python
df.columns = df.columns.str.strip()

# Drop missing values if any
df = df.dropna()

print("Cleaned Data Shape:", df.shape)
```

```python
print(df.columns)
```

```python
import matplotlib.pyplot as plt

# Clean column names
df.columns = df.columns.str.strip()

# Group by state using correct column names
state_sales = df.groupby('State Name')['Grand Total'].sum().sort_values(ascending=False)

plt.figure()
state_sales.head(10).plot(kind='bar')
plt.title("Top 10 EV Sales by State")
plt.xlabel("State")
plt.ylabel("Total EV Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
vehicle_cols = [
    'Two Wheeler', 'Three Wheeler', 'Four Wheeler',
    'Goods Vehicles', 'Public Service Vehicle',
    'Special Category Vehicles', 'Ambulance/Hearses',
    'Construction Equipment Vehicle', 'Other'
]
```

```python
# Sort vehicle sales
vehicle_sales = df[vehicle_cols].sum().sort_values(ascending=False)

# Take top 3
top3 = vehicle_sales.head(3)

# Combine rest as "Others"
others = vehicle_sales[3:].sum()

# Create new series
final_data = top3.copy()
final_data['Others'] = others

# Plot clean pie chart
import matplotlib.pyplot as plt

plt.figure()
final_data.plot(kind='pie', autopct='%1.1f%%')
plt.title("EV Sales Distribution (Top Categories)")
plt.ylabel("")
plt.show()
```

```python
charging = df.groupby('State Name')['total-charging-stations'].sum().sort_values(ascending=False)

plt.figure()
charging.head(10).plot(kind='bar')
plt.title("Top 10 States by Charging Stations")
plt.xlabel("State")
plt.ylabel("Charging Stations")
plt.xticks(rotation=45)
plt.show()
```

## **📊 Insights**

- Three-wheelers dominate EV sales (~67%), showing strong commercial adoption  
- Two-wheelers (~29%) indicate affordability-driven growth  
- Four-wheelers have lower adoption but growing steadily  
- States with higher charging infrastructure show higher EV penetration  

## 📌 Conclusion
EV adoption in India is driven by affordability (2W/3W) and infrastructure availability. Strategic expansion of charging networks can significantly accelerate market growth.
